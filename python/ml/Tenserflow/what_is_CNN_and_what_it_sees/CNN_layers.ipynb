```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
import cv2

import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
```

```python
model = load_model('../image_classifier/imageclassifier.h5')
```

```python
len(model.layers)
```

```python
model.layers[:6]
```

```python
inter_layers = model.layers[:3]
```

```python
inter_model = Sequential(inter_layers)
```

```python
inter_model.input_shape
```

```python
img = cv2.imread('../image_classifier/8iAb9k4aT.jpg')
resize = cv2.resize(img, (256, 256))
```

```python
pred = inter_model.predict(np.expand_dims(resize, axis=0))
```

```python
pred.shape
```

```python
num_rows = 4
num_cols = 4

fig, ax = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(10, 10))
for idx, ax_idx in enumerate(list(itertools.product(range(num_cols), range(num_rows)))):
  ax[ax_idx[1], ax_idx[0]].imshow(pred[0, :, :, idx])
```
